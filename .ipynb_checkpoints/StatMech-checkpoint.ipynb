### 1. The nature of statistical mechanics
Statistical Mechanics is useful for systems with degrees of freedom that are either too large to track, or too difficult to measure precisely. Generally, deductive (Bayesian) or inductive (Frequentist) approaches can be used. Deductive means making assumptions to simplify the problem, and inductive involves observing repeated trials
